```python
import pandas as pd
import yfinance as yf
import mplfinance as mpf
import numpy as np
# Documentation: https://ranaroussi.github.io/yfinance/index.html
```

```python
def ticker_to_isin(t : str) -> str:
  ticker = yf.Ticker(ticker= t)
  return ticker.get_isin()

aapl_isin = ticker_to_isin("AAPL")
```

```python
yf.Ticker(ticker = 'AAPL').history(period="1mo", auto_adjust=False, actions=False)
```

```python
def data_analysis(t : str):
  ticker = yf.Ticker(ticker = t)
  
  history = ticker.history(period="1mo", auto_adjust=False, actions=False)
  mpf.plot(history, type='candle', style='charles', title='AAPL Candlestick Chart')


data_analysis("AAPL")
```
